```python
#como todas as fun√ß√µes s√£o definidas em rela√ß√£o ao t, vamos definir o t
step = 0.1 
t = np.arange(0, 5+step, step)
t
```

```python
#p/: alpha > w0 --> amortecimento supercr√≠tico
#i(t) = A1 * e^(s1 * t) + A2 * e^(s2 * t)

#Ent√£o vamos supor os valores dessas constantes:
alpha = 3
w0 = 2.5

A1 = 80
A2 = -70

s1 = -alpha + mt.sqrt((alpha**2) - (w0**2))
s2 = -alpha - mt.sqrt((alpha**2) - (w0**2))

#vamos criar um i cheio de zeros para o amortecimento supercritico

i_amort_supercri = np.zeros(len(t))

for j in range(len(t)):
  i_amort_supercri[j] = A1 * mt.exp(s1 * t[j]) + A2 * mt.exp(s2 * t[j])

plt.plot(t, i_amort_supercri, 'b')
plt.show()
```

```python
#p/: alpha = w0 --> amortecimento cr√≠tico
#i(t) = A1 * t * exp(s * t) + A2 * exp(s * t)

#Ent√£o vamos supor os valores dessas constantes:
alpha = 3
w0 = 3

A1 = 80
A2 = 5

s = -alpha

#vamos criar um i cheio de zeros para o amortecimento supercritico

i_amort_cri = np.zeros(len(t))

for j in range(len(t)):
  i_amort_cri[j] = A1 * t[j] * mt.exp(s * t[j]) + A2 * mt.exp(s * t[j])

plt.plot(t, i_amort_cri, 'g')
plt.show()
```

```python
#p/: alpha < w0 --> subamortecimento
#i(t) = exp(-alpha * t) * (B1 * cos(wd * t) + B2 * sin(wd * t))

#Ent√£o vamos supor os valores dessas constantes:
alpha = 1.5
w0 = 8

B1 = 10
B2 = 30

wd = mt.sqrt((w0**2) - (alpha**2))

#vamos criar um i cheio de zeros para o amortecimento supercritico

i_sub_amort = np.zeros(len(t))

for j in range(len(t)):
  i_sub_amort[j] = mt.exp(-alpha * t[j]) * (B1 * mt.cos(wd * t[j]) * (B2 * mt.sin(wd * t[j])))

plt.plot(t, i_sub_amort, 'r')
plt.show()
```

```python
#agora vamos deixar os gr√°ficos juntos e bonitos

plt.figure(figsize=(10, 5))

plt.plot(t, i_amort_supercri, 'b', linestyle = 'dotted')
plt.plot(t, i_amort_cri, 'g', linestyle = 'dashed')
plt.plot(t, i_sub_amort, 'r')

plt.xlabel('t [s]', fontsize = 20)
plt.ylabel('i(i) [A]', rotation=0, fontsize = 20)
plt.title('Gr√°fico  i x t', fontsize = 30)
plt.grid(axis = 'y')

plt.legend(["Superamortecida (Œ± > œâ0)", "Criticamente Amortecida (Œ± = œâ0)", "Subamortecida (Œ± < œâ0)"], loc ="upper right")
plt.show()
```
